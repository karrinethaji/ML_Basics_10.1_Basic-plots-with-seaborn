# Plots using seaborn library

Seaborn is a data visualization library in Python that is built on top of Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn comes with several built-in functions for creating various types of plots. Here are some common types of plots available in Seaborn:

- Scatter Plot:

scatterplot(): Displays the relationship between two numeric variables by plotting points.

- Line Plot:

lineplot(): Draws a line connecting observations in the dataset, useful for visualizing trends over time.

- Bar Plot:

barplot(): Shows the relationship between a categorical variable and a numeric variable using rectangular bars.

- Count Plot:

countplot(): Displays the counts of observations in each category of a categorical variable.

- Box Plot:

boxplot(): Illustrates the distribution of a quantitative variable (numerical) across different levels of one or more categorical variables.

A quantitative variable is a variable which can have some numerical value i.e. it can be represented in numbers

- Violin Plot:

violinplot(): Combines aspects of a box plot and a kernel density plot, providing information about the distribution of the data.

- Heatmap:

heatmap(): Displays the relationships between two categorical variables or a matrix of data using colors.

- Pair Plot:

pairplot(): Creates a grid of scatterplots for each pair of numerical features in a dataset, along with histograms on the diagonal.

- Joint Plot:

jointplot(): Combines scatter plots for two numerical variables and histograms for each individual variable.

- Facet Grid:

FacetGrid(): Allows you to create a grid of subplots based on the levels of one or more categorical variables.

- Regression Plot:

regplot(): Combines a scatter plot and a regression line to show the relationship between two numeric variables.

- LM Plot:

lmplot(): Similar to regplot(), but designed to work well with FacetGrid for visualizing relationships across different subsets of the data.

These are just a few examples, and Seaborn provides many more options and customization features for creating visually appealing and informative plots. You can refer to the official Seaborn documentation for more details and examples: https://seaborn.pydata.org/

**Scatter plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.scatterplot(x="total_bill", y="tip", data=tips)

plt.show()
```

```python
tips[["total_bill","tip"]].corr()
```

**Line Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.lineplot(x="total_bill", y="tip", data=tips)

plt.show()
```

**Heat Map**

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")
print(flights)
pivot_flights = flights.pivot_table(index="month", columns="year", values="passengers")
print(pivot_flights)
plt.figure(figsize=(20,6))
sns.heatmap(pivot_flights,annot=True)
plt.show()
```

**Pair Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
sns.pairplot(iris)
plt.show()
```

**Joint Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.jointplot(x="total_bill", y="tip", data=tips,hue="sex")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.jointplot(x="total_bill", y="tip", data=tips,kind="reg")
plt.show()
```

**Facet Grid**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
g = sns.FacetGrid(tips, col="time", row="sex")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()
```

**Regression Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.regplot(x="total_bill", y="tip", data=tips)
plt.show()

# we can observe the below plot is same as joint plot if we initialize the kind="reg" in the joint plot

# disadvantage is that we can't able to use "hue" feature, so we have LM plot to limit this disadvantage
```

**LM Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.lmplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.show()
```
